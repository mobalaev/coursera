```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('weights_heights.csv', index_col='Index')
```

```python
data.plot(y='Height', kind='hist', 
           color='red',  title='Height (inch.) distribution');
```

```python
data.head()
```

```python
data.plot(y='Weight', kind='hist', 
           color='green',  title='Weight(Pounds)  distribution');
```

```python
def make_bmi(height_inch, weight_pound):
    METER_TO_INCH, KILO_TO_POUND = 39.37, 2.20462
    return (weight_pound / KILO_TO_POUND) / \
           (height_inch / METER_TO_INCH) ** 2
```

```python
data['BMI'] = data.apply(lambda row: make_bmi(row['Height'], 
                                              row['Weight']), axis=1)
```

```python
sns.pairplot(data, palette="husl",diag_kind="kde",kind="reg");
```

```python
def weight_category(weight):
    category_type=0
    if weight<120:
            category_type=1
    elif weight>=150:
            category_type=3
    else:
            category_type=2
    return category_type

data['weight_cat'] = data['Weight'].apply(weight_category)
sns.boxplot(x=data['weight_cat'],y=data['Height']);
data.head()
```

```python
data.plot(x='Height', y='Weight', title='Correlation between weight and height', kind='scatter');
```

```python
def measurement_error(w1, w0, data):
    summa = 0.0
    for _, row in data.iterrows():
        summa+=(row['Height'] - (w0 + w1 * row['Weight'])) ** 2
    return summa
```

```python
def y1(x):
    return 60 + 0.05 * x

def y2(x):
    return 50 + 0.16 * x

x = np.linspace(60, 170)
data.plot(x='Weight', y='Height', kind='scatter', title='Correlation between weight and height')
plt.plot(x, y1(x),color='green',)
plt.plot(x, y2(x),color='red')
plt.legend( ('line (60, 0.05)', 'line (50, 0.16)') );
```

```python
w0 = 50.0
error_graph = []
for w1 in np.arange(-5.0, 5.0, 0.25):
    error_graph.append(measurement_error(w1, w0, data))
    plt.plot(error_graph);
```

```python
import scipy.optimize
from scipy.optimize import minimize_scalar

w0 = 50
w1_opt = minimize_scalar(measurement_error, bounds=(-5, 5), args=(w0,data))
print (w1_opt)
```

```python
def y(x):
    w0 = 50
    return w0 + w1_opt.x * x
    
x = np.linspace(75, 180)
data.plot(x='Weight', y='Height', kind='scatter', title='Minimize_scalar method')
plt.plot(x, y(x),color='red');
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

#   NumPy      X  . 
#   meshgrid,      
#   .    Z(x, y).
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# ,   *plot_surface*  
#  Axes3DSubplot.   .
surf = ax.plot_surface(X, Y, Z)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show();
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

w0 = np.arange(-5, 5, 0.25)
w1 = np.arange(-5, 5, 0.25)
w0, w1 = np.meshgrid(w0, w1)
Z = measurement_error(w1, w0,data)

surf = ax.plot_surface(w0, w1, Z)
ax.set_xlabel('Intercept')
ax.set_ylabel('Slope')
ax.set_zlabel('Error')
ax.set_title('Measurement Error in 3D')
plt.show();
```

```python
from scipy.optimize import minimize
    
def yyy(w):
    return measurement_error(w[1], w[0],data)
    
w_opt = minimize(yyy, [0.0, 0.0], method='L-BFGS-B', bounds=([-100, 100],[-5, 5]))


print (w_opt)
```

```python
w0_opt = w_opt.x[0]
w1_opt = w_opt.x[1]
    
def y(x):
    return w0_opt + w1_opt * x
    
x = np.linspace(60, 200)
data.plot(x='Weight', y='Height', kind='scatter', title='Method L-BFGS-B minimizer')
plt.plot(x, y(x),color='red');
```
